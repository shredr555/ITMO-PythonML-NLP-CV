# Задание 3


1. С помощью библиотеки OpenCV считайте изображение
2. Сожмите изображение до ширины 890, соблюдая пропорции исходного изображения. Важно: коэффициент отношения сторон при вычислениях не округляйте. Если высота не окажется целой, округлите значение до целого в меньшую сторону.
3. Сохраните сжатое изображение в палитре RGB.



```python
import matplotlib.pyplot as plt
%matplotlib inline

#1.
bgr_img = cv2.imread('3.jpg')

#2.
h, w, c = bgr_img.shape
new_width = 890

k = float(w) / h
print(k)
dim = (new_width, int(new_width / k))

resized_BGR_img = cv2.resize(bgr_img, dim, interpolation = cv2.INTER_AREA)
plt.imshow(resized_BGR_img)
plt.show()

#3.
h, w, c = resized_BGR_img.shape
print("высота h", h)

#3_3
file_name = 'result.jpg'
cv2.imwrite(file_name, resized_BGR_img)

import os

print("размер файла в байтах", os.path.getsize(file_name))
```
